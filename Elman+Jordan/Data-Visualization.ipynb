```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
file_path = 'dataset/dataset.csv'
dataset = pd.read_csv(file_path)
dataset.head()
```

```python
summary_statistics = dataset.describe()
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
sns.histplot(dataset['high'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of High Prices')

sns.histplot(dataset['low'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Low Prices')

sns.histplot(dataset['open'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Open Prices')

sns.histplot(dataset['close'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Close Prices')

sns.histplot(dataset['volume'], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Distribution of Volume')

axes[2, 1].axis('off')

plt.tight_layout()
plt.savefig("report/before_preparing.png")
plt.show()
```

```python
file_path = 'dataset/cleaned_dataset.csv'
cleaned_dataset = pd.read_csv(file_path)
cleaned_dataset.head()
cleaned_dataset.fillna(0)
cleaned_dataset.set_index('jdate', inplace=True)
```

```python
summary_statistics = cleaned_dataset.describe()
fig, axes = plt.subplots(4, 2, figsize=(15, 15))
sns.histplot(cleaned_dataset['high'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of High Prices')

sns.histplot(cleaned_dataset['low'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Low Prices')

sns.histplot(cleaned_dataset['open'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Open Prices')

sns.histplot(cleaned_dataset['close'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Close Prices')

sns.histplot(cleaned_dataset['volume'], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Distribution of Volume')


sns.histplot(cleaned_dataset['returns'], bins=30, kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Distribution of return')

sns.histplot(cleaned_dataset['volume_change'], bins=30, kde=True, ax=axes[3, 0])
axes[3, 0].set_title('Distribution of Volume Change')



axes[3, 1].axis('off')

plt.tight_layout()
plt.savefig("report/after_preparing.png")

plt.show()
```
